```python
%%script scala
Encapsulation -> hiding data using private vars + getters/setters
class BankAccount(private var balance: Double) {
  // getter
  def getBalance: Double = balance

  // setter with validation
  def deposit(amount: Double): Unit = {
    if (amount > 0) balance += amount
    else println("Invalid deposit")
  }

  def withdraw(amount: Double): Unit = {
    if (amount > 0 && amount <= balance) balance -= amount
    else println("Invalid withdraw")
  }
}
// Demo
object EncapsulationDemo {
  def main(args: Array[String]): Unit = {
    val acc = new BankAccount(1000)
    acc.deposit(500)
    acc.withdraw(200)
    println("Final Balance: " + acc.getBalance)
  }
}

EncapsulationDemo.main(Array())
```

```python
## part 2 :  Inheritance and overiding
%%script scala
// part 2 : Inheritance and overiding
// Parent class
class Person(val name: String, val age: Int) {
  def showinfop(): Unit = println("Name: name,Age: age")
  }
// child class extending Person
class Employee(name: String, age: Int, val role: String) extends Person(name, age) {
  override def showinfop(): Unit = println("Name: " + name + ", Age: " + age + ", Role: " + role)
}

// demo
object InheritanceDemo {
  def main(args: Array[String]): Unit = {
    val emp = new Employee("John", 30, "Manager")
    emp.showinfop()
  }
}
InheritanceDemo.main(Array())
```

```python
# Part 3: Abstraction

%%script scala
// =========================================
// Part 3: Abstraction
// =========================================

// Abstract class
abstract class Shape {
  def area(): Double
}

// Implementations
class Circle(val radius: Double) extends Shape {
  def area(): Double = Math.PI * radius * radius
}
class Rectangle(val length: Double, val width: Double) extends Shape {
  def area(): Double = length * width
}

// Demo
object AbstractionDemo {
  def main(args: Array[String]): Unit = {
    val shapes: List[Shape] = List(new Circle(5), new Rectangle(4, 6))
    shapes.foreach(s => println(s"Area: ${s.area()}"))
  }
}

AbstractionDemo.main(Array())# Part 3: Abstraction

%%script scala
// =========================================
// Part 3: Abstraction
// =========================================

// Abstract class
abstract class Shape {
  def area(): Double
}

// Implementations
class Circle(val radius: Double) extends Shape {
  def area(): Double = Math.PI * radius * radius
}
class Rectangle(val length: Double, val width: Double) extends Shape {
  def area(): Double = length * width
}

// Demo
object AbstractionDemo {
  def main(args: Array[String]): Unit = {
    val shapes: List[Shape] = List(new Circle(5), new Rectangle(4, 6))
    shapes.foreach(s => println(s"Area: ${s.area()}"))
  }
}

AbstractionDemo.main(Array())
```

```python
# Part 4: Polymorphism (Overloading & Overriding)


%%script scala
// =========================================
// Part 4: Polymorphism
// =========================================

class Calculator {
  // Method Overloading
  def add(a: Int, b: Int): Int = a + b
  def add(a: Double, b: Double): Double = a + b
}

class AdvancedCalculator extends Calculator {
  // Method Overriding
  override def add(a: Int, b: Int): Int = {
    println("Using AdvancedCalculator logic")
    a + b + 10
  }
}

// Demo
object PolymorphismDemo {
  def main(args: Array[String]): Unit = {
    val calc = new Calculator
    println("Int Add: " + calc.add(3, 7))
    println("Double Add: " + calc.add(3.5, 2.5))

    val advCalc: Calculator = new AdvancedCalculator
    println("Overridden Add: " + advCalc.add(3, 7))
  }
}

PolymorphismDemo.main(Array())
```

```python
## part 5 : traits (multiple inheritance)
%%script scala
// part 5: traits
traits Printable {
  def printInfo(): Unit
  }
  trait Idenitfiable{
    def id: String
  }

  // Class using multiple traits
  class Student(val id: String, val name: String) extends Printable with Idenitfiable {
    override def printInfo(): Unit = println("ID: " + id + ", Name: " + name)
  }

// demo
objectTraitDemo
def main(args: Array[String]): Unit = {
  val student = new Student("123", "John")
  student.printInfo()
}
TraitDemo.main(Array())

```

```python

# Part 5: Traits (Multiple Inheritance)


%%script scala
// =========================================
// Part 5: Traits
// =========================================

trait Printable {
  def printInfo(): Unit
}
trait Identifiable {
  def id: String
}

// Class using multiple traits
class Student(val id: String, val name: String) extends Printable with Identifiable {
  def printInfo(): Unit = println(s"Student -> ID:
name")
}

// Demo
object TraitDemo {
  def main(args: Array[String]): Unit = {
    val student = new Student("S101", "Bob")
    student.printInfo()
  }
}

TraitDemo.main(Array())
```
